```python
import matplotlib.pyplot as plt

ages = [21, 25, 30, 35, 40, 45, 50, 55, 60, 65]
population_count = [150, 200, 300, 400, 350, 250, 200, 150, 100, 50]

plt.bar(ages, population_count, color='blue')

plt.xlabel('Age')
plt.ylabel('Population Count')
plt.title('Distribution of Ages in a Population')

plt.show()
```
